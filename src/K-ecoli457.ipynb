```python
Es, Ms = [], []
for i, row in df.iterrows():
    E1 = row["Estimated range"]; E2 = row["Unnamed: 6"]
    M1 = row["Measured range"]; M2 = row["Unnamed: 8"]
    Es.append([E1, (E1+E2)/2, E2])
    Ms.append([M1, (M1+M2)/2, M2])
```

```python
X = np.log10([E[1] for E in Es])
Y = np.log10([M[1] for M in Ms])
plt.scatter(X, Y, s=5)
plt.plot([-3, 2], [-3, 2], '--',color="black", alpha=0.5)
plt.xlim((-3, 2)); plt.ylim((-3, 2))
plt.show()
```

```python
ALL_V_UP = []
ALL_V_DN = []
ALL_M_UP = []
ALL_M_DN = []
for i in range(457):
    idx = i+1
    V1,V2,V3,M1,M2,M3 = PermuteRNX(idx)
    V_cor_up, p1 = pearsonr(V1, V3); ALL_V_UP.append(V_cor_up)
    V_cor_dn, p2 = pearsonr(V1, V2); ALL_V_DN.append(V_cor_dn)
    M_cor_up, p3 = pearsonr(M1, M3); ALL_M_UP.append(M_cor_up)
    M_cor_dn, p4 = pearsonr(M1, M2); ALL_M_DN.append(M_cor_dn)
```

```python
fig, axs = plt.subplots(2, 2, dpi=120)
axs[0, 0].hist(ALL_V_UP)
axs[0, 0].set_title('2 vs 1 Vnet')
axs[0, 1].hist(ALL_V_DN)
axs[0, 1].set_title('0.5 vs 1 Vnet')
axs[1, 0].hist(ALL_M_UP)
axs[1, 0].set_title('2 vs 1 Concentration')
axs[1, 1].hist(ALL_M_DN)
axs[1, 1].set_title('0.5 vs 1 Concentration')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```
